```python
column_list = airbnb.columns.tolist()
#print(column_list)
for i in column_list:
    print (f"Value counts for {i}:")
    print(airbnb[i].value_counts())
    #print(airbnb[i])
    print("\n")
```

```python
# Check for missing values
missing_values = airbnb.isnull().sum()
print(missing_values)
```

```python
# Check missing values for Beds column
missing_in_beds = airbnb[airbnb['Beds'].isnull()]
print(missing_in_beds)
```

```python
# Filter rows that are private room to check how many beds they have
filtered_rows = airbnb[airbnb['RoomType'] == 'Private room']

# Get the count of records
count = filtered_rows.shape[0]
print(f"Count of records with 'Private Room' in RoomType: {count}")

# Print filtered records
print(filtered_rows)
```
